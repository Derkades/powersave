```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv('measurements.csv')
```

```python
for value1 in [11, 25, 33 , 15, 35, 77, 27, 16]:
    column = 'partition point beide' 
    newdata = data[data[column] == value1]
    for value2 in [500000, 1398000, 1800000, 2208000]:
        graphdata = newdata[newdata['Big Freq'] == value2]
        x = pd.Series.tolist(graphdata['Little Freq'])
        y = pd.Series.tolist(graphdata['Watt'])

        plt.figure()
        plt.title(f"Big Freq = {value2},  Partipoint = {value1}")
        plt.xlabel("Little Freq")
        plt.ylabel("Watt")
        plt.plot(x,y)
        plt.show()
```

```python
big_val_list = [500000, 1398000, 1800000, 2208000]
little_val_list = [500000, 1398000, 1800000]
for value1 in [11, 25, 33 , 15, 35, 77, 27, 16]:
    matrix = []
    column = 'partition point beide' 
    newdata = data[data[column] == value1]
    for big_val in big_val_list:
        matrix_row = []
        for little_val in little_val_list:
            graphdata = newdata[newdata['Big Freq'] == big_val]
            graphdata = graphdata[graphdata['Little Freq'] == little_val]
            y = pd.Series.tolist(graphdata['Watt'])[0]
            matrix_row.append(y)
        matrix.append(matrix_row)
    
    matrix = np.array(matrix)
    maxnum = np.max(matrix)
    matrix = matrix / maxnum
    plt.figure(figsize=(8,8))
    plt.title(f"Power uage for partition point {value1}", fontsize=15)
    plt.xlabel("Little Frequency", fontsize=15)
    plt.ylabel("Big Frequency", fontsize=15)
    plt.xticks(np.arange(len(little_val_list)), little_val_list, rotation=90)
    plt.yticks(np.arange(len(big_val_list)), big_val_list)

    colormap = plt.imshow(matrix)
    plt.colorbar(colormap)
    plt.show()
```
